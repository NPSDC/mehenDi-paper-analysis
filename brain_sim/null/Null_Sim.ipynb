#### Path to null simulations
We have a total of 10 null simulations, where we have already created the TreeSummarizedExperiment object with beaveR

```r
suppressPackageStartupMessages(library(glue))
suppressPackageStartupMessages(library(mehenDi))
suppressPackageStartupMessages(library(TreeSummarizedExperiment))
suppressPackageStartupMessages(library(phangorn))
simDir <- "/fs/cbcb-scratch/jfan03/treeDE-null"
seedFiles <- sapply(seq(10), function(seed) {
    file.path(simDir, glue(paste("output_seed","",sep="="), {seed}),
             paste(glue("treeSumExp_seed=", {seed}), "RData", sep="."))
})
```

```r
nl <- c() ## Number of transcripts
innNodes <- c() ## Number of inner nodes
keptL <- c() ### Number of transcripts with labelKeep=T
keptInn <- c() ### Number of inner nodes with labelKeep=T
pInn <- c()
pAll <- c()
pL <- c()
pSep <- c() ## All pvalues but leaf and inner computed separately
tDL <- list() ### mehenDi RUns
test_stat <- c()
alphas <- 10^seq(-9,-1,1)
fpsmehenDi <- vector(mode="list", length(seedFiles))
fprsmehenDi <- vector(mode="list", length(seedFiles))
fpsTxp <- vector(mode="list", length(seedFiles))
fprsTxp <- vector(mode="list", length(seedFiles))
for(i in seq_along(seedFiles)) {
    load(seedFiles[i])
    tree <- rowTree(tse)
    l <- length(tree$tip.label)
    print(paste(".......Null Simulation", i, "............"))
    
    yAll <- beaveR::computeSizeFactors(tse)
    yAll <- beaveR::scInfReps(yAll)
    yAll <- fishpond::labelKeep(yAll)
    metadata(yAll)[["infRepsScaled"]] <- TRUE
    nl <- c(nl, l)
    innNodes <- c(innNodes, length((l+1):nrow(yAll)))
    keptL <- c(keptL, sum(mcols(yAll)[["keep"]][1:l]))
    keptInn <- c(keptInn, sum(mcols(yAll)[["keep"]][(l+1):nrow(yAll)]))
    
    print("Swish run on all nodes together")
    set.seed(10)
    yAll <- fishpond::swish(yAll, x="condition")
    pAll <- c(pAll, mcols(yAll)[["pvalue"]])
    pInn <- c(pInn, mcols(yAll)[["pvalue"]][(l+1):nrow(yAll)])
    pL <- c(pL, mcols(yAll)[["pvalue"]][1:l])
    print("All Nodes")
#     hist(mcols(yAll)[["pvalue"]])
    print("Leaf Nodes")
#     hist(mcols(yAll)[1:l,"pvalue"])
    print("Inner Nodes")
#     hist(mcols(yAll)[(l+1):nrow(yAll),"pvalue"])
    
    print("Swish run separately")
    set.seed(10)
    yTxps <- fishpond::swish(yAll[1:l,], x="condition")
    print("Only on the leaf")
    hist(mcols(yTxps)[["pvalue"]])
    set.seed(10)
    yInn <- fishpond::swish(yAll[(l+1):nrow(yAll),], x="condition")
    print("Only on the inner nodes")
    hist(mcols(yInn)[["pvalue"]])
    
                 
    print("Running mehenDi")
    pvals <- c(mcols(yTxps)[["pvalue"]], mcols(yInn)[["pvalue"]])
    pSep <- c(pSep, pvals)
    print("length of selected nodes output by mehenDi")
    tDL[[i]] <- lapply(alphas, function(alpha) {
        mehenDi(yAll, x="condition", pvals, minP=0.70, mIRVThresh = 0.40, pThresh = alpha, cores=4)
    })
    
    fpsmehenDi[[i]] <- sapply(tDL[[i]], function(x) length(x[["selNodes"]]))
    print(fpsmehenDi[[i]])
    fprsmehenDi[[i]] <- sapply(tDL[[i]], function(x) {
        cN <- x[["selNodes"]]
        fpr <- 0
        if(length(cN) > 0) {
            rN <- setdiff(seq(l), unlist(Descendants(tree, cN)))
            tN <- union(cN, rN) ## true negatives
            fpr <- length(cN)/(length(cN) + length(tN))
        }
        fpr
    })
    
    print("Number of differentially expressed transcripts")
    fpsTxp[[i]] <- sapply(alphas, function(alpha) sum(mcols(yTxps)[["pvalue"]] < alpha, na.rm=T))
    print(fpsTxp[[i]])                          
    fprsTxp[[i]] <- sapply(seq_along(alphas), function(j) fpsTxp[[i]][j]/(fpsTxp[[i]][j]+l))

    yAll <- fishpond::swish(yAll, x="condition", returnNulls = TRUE)
    test_stat <- c(test_stat, yAll$stat)
}
```

```r
save(nl, file = "nl.RData")
save(innNodes, file = "innNodes.RData")
save(keptL, file = "keptL.RData")
save(keptInn, file = "keptInn.RData")
save(pInn, file = "pInn.RData")
save(pAll, file = "pAll.RData")
save(pL, file = "pL.RData")
save(fpsmehenDi, file = "fpsmehenDi.RData")
save(fprsmehenDi, file = "fprsmehenDi.RData")
save(fpsTxp, file = "fpsTxp.RData")
save(fprsTxp, file = "fprsTxp.RData")
save(test_stat, file = "test_stat.RData")
save(tDL, file = "tDL.RData")
```

#### False positive rate

```r
dfFPRs <- data.frame(fprs = rep(0, length(fprsmehenDi)*length(fprsmehenDi[[1]])), 
                     pthresh=rep(10^(seq(-9,-1, 1)),length(fprsmehenDi)))
for(i in seq_along(fprsmehenDi))
    dfFPRs[["fprs"]][(i-1)*length(fprsmehenDi[[i]])+(1:length(fprsmehenDi[[i]]))] <- fprsmehenDi[[i]]

dfFPRs <- dfFPRs %>% group_by(pthresh) %>%
            summarize(mean_fpr = mean(fprs), sd_fpr = sd(fprs))
dfFPRs
```

```r
x_min <- floor(log10(min(dfFPRs$pthresh)))
x_max <- ceiling(log10(max(dfFPRs$pthresh)))
size <- 28
pmehenDi <- ggplot(dfFPRs, aes(x = pthresh, y = mean_fpr)) +
  geom_line(color = "blue") +
  geom_errorbar(aes(ymin = mean_fpr - sd_fpr, ymax = mean_fpr + sd_fpr), width = 0.5) +
    geom_point(size=2) +
    scale_x_log10(limits=c(-1e-10,0.5), expand=c(0,0)) +
    scale_y_log10(limits=c(-1e-6,1e-1)) +
    theme_bw() +
    theme(title=element_text(size=size-1), 
         axis.text=element_text(size=size+8),
        axis.title=element_text(size=size+8)) +
  labs(title = "FPR for mehenDi at different p-value \nthresholds", x = "p-value", y = "FPR")
```
